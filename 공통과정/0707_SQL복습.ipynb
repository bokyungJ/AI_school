```python
# 방법 1

import sqlite3

dbpath=r"c:\Apps\database\world.db"

conn = sqlite3.connect(dbpath) # 신규생성 또는 접속
cur = conn.cursor() # 커서객체
cur.execute("select * from country limit 5")

rows = cur.fetchall()
for row in rows: # 한 행씩 값 가져오기, 튜플로 가져옴
    print(row)

conn.commit() # 확인
conn.close() # 데이터베이스 연결 종룔
```

```python
# 방법 3 : 순서에 맞춰서 하는 파라미터 방식 (튜플)
# - ?를 사용하면 조건을 그때그때 바꿀 수 있음

import sqlite3

dbpath=r"c:\Apps\database\world.db"

conn = sqlite3.connect(dbpath)
cur = conn.cursor()
sql = "select name, population as p from country where p between ? and ?"
cur.execute(sql,('5000','10000'))

rows = cur.fetchall()
for row in rows:
    print(row)

conn.commit()
conn.close()
```

```python
# 방법 4 : 딕셔너리 형태로 파라미터 사용하기

import sqlite3

dbpath=r"c:\Apps\database\world.db"

conn = sqlite3.connect(dbpath)
cur = conn.cursor()
sql = "select name, population as p from country where p between :s and :e"
cur.execute(sql,{"s":"5000","e":"10000"})

rows = cur.fetchall()
for row in rows:
    print(row)
    
conn.commit()
conn.close()
```

```python
# 함수로 만들기

import sqlite3

dbpath = r"c:\Apps\database\world.db"

def printPopuRange(start, end):
    conn = sqlite3.connect(dbpath)
    cur = conn.cursor()
    sql = "select name, population as p from country where p>:s and p<:e"
    cur.execute(sql,{"s":start,"e":end})
    
    rows = cur.fetchall()
    for row in rows:
        print(row)
        
    conn.commit()
    conn.close()

printPopuRange(5000,10000)
```
